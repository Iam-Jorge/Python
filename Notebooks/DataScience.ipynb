## 1.Introducción a la Ciencia de datos <a class="anchor" id="t1"></a>
![title](../img/Divider.png)

El término "Ciencia de Datos" se acuñó a principios del siglo XXI. Se atribuye a William S. Cleveland, que en 2001 escribió "Data Science: An Action Plan for Expanding the Technical Areas of the Field of Statistics". Aproximadamente un año después, el International Council for Science: Committee on Data for Science and Technology comenzó a publicar la revista CODATA Data Science Journal a partir de abril de 2002. Poco después, en enero de 2003, la Universidad de Columbia empezó a publicar The Journal of Data Science.


**Desencadenante**

Durante los años 1998-2000, los discos duros se abarataron mucho. Las empresas y los gobiernos empezaron a comprar muchos. 

>Un corolario de la Ley de Parkinson es que los datos siempre se expanden para llenar el espacio disponible en el disco.

La interacción "disco-datos" es un ciclo exponencial positivo entre la compra de cada vez más discos y la acumulación de cada vez más datos. Este ciclo produce big data. Big data es un término utilizado para describir conjuntos de datos tan grandes y complejos que resulta difícil trabajar con ellos utilizando las herramientas habituales de gestión de bases de datos.


**Caso práctico**

A mediados y finales de los 90, AltaVista era el motor de búsqueda más popular de Internet. Enviaba "crawlers o rastreadores" para extraer el texto de todas las páginas de la web. Los crawlers devolvían el texto a AltaVista. AltaVista indexaba todo el texto. Así, cuando una persona buscaba una palabra clave, AltaVista podía encontrar las páginas web que contenían esa palabra. AltaVista presentaba entonces los resultados como una lista ordenada de páginas web, con las páginas que tenían las menciones más frecuentes del término en la parte superior.

A finales de los 90, los fundadores de Google inventaron una forma diferente de hacer búsquedas. Combinaron matemáticas, estadística, ingeniería de datos, computación avanzada para crear un motor de búsqueda que desplazó a AltaVista. El algoritmo se conoce como PageRank. PageRank no sólo tiene en cuenta las palabras de la página, sino también los hipervínculos. PageRank asume que un hipervínculo entrante es un indicador de que alguna otra persona pensó que la página actual era lo suficientemente importante como para poner un enlace a ella en su propia página. Así, las páginas con más hipervínculos acaban encabezando la lista de resultados de búsqueda. PageRank captura el conocimiento humano sobre las páginas web, además del contenido.

Google no sólo rastreó la web, sino que la ingirió. Eso es big data. Luego tienen que calcular el algoritmo PageRank a través de esos grandes datos. Eso requiere un cálculo masivo. Luego tienen que hacer que la búsqueda y los resultados de búsqueda sean rápidos para todo el mundo. La búsqueda de Google es un triunfo de la ciencia de datos (aunque no se llamaba ciencia de datos cuando empezó).

### 1.1. ¿Qué son los datos? <a class="anchor" id="t1_1"></a>

La palabra "datos" es una palabra de uso general que denota una colección de observaciones que se miden y comunican de forma que resulten inteligibles tanto para el que los registra (una persona o dispositivo) como para el que los lee.

¿Qué es un punto de datos?
- Un punto de datos es una unidad discreta de información. En un sentido general, cualquier hecho aislado es un punto de datos.

¿Qué es un conjunto de datos?
- Es una colección de puntos de datos que han sido observados en objetos similares y formateados de manera similar.

¿Qué son los tipos de datos?
- Los datos pueden existir en muchas formas, como texto, números, imágenes, audio y vídeo.

### 1.2. ¿Qué es la ciencia de datos? <a class="anchor" id="t1_2"></a>

La ciencia de datos es una mezcla de varias disciplinas diferentes que estudia cómo se pueden recopilar, almacenar y manejar los datos utilizando ordenadores. Mientras que la estadística nos proporciona fundamentos matemáticos, la ciencia de datos aplica conocimientos y conceptos matemáticos para extraer conclusiones y perspectivas a partir de una amplia gama de ámbitos de dominio.


**Método científico**

Es la parte científica de la ciencia de datos. Es un proceso para adquirir nuevos conocimientos aplicando los principios del razonamiento a las pruebas empíricas derivadas de la comprobación de hipótesis mediante experimentos repetibles.

*Principios de razonamiento*

Existen dos formas generales de razonamiento lógico: inductivo y deductivo. Wl razonamiento inductivo llega a principios generales a partir de observaciones específicas, mientras que el razonamiento deductivo llega a conclusiones específicas basadas en principios generales.

Algunos de los paradigma de la ciencia:
- Pruebas empíricas: Son datos obtenidos mediante la observación o la experimentación. Esto contrasta con los datos derivados de argumentos lógicos o conclusiones propagadas por mitos y leyendas.
- Comprobación de hipótesis: Este proceso suele afirmar dos proposiciones, de las cuales sólo una puede ser cierta.
- Experimentos repetibles: Es un procedimiento metódico de ensayo y error llevado a cabo con el objetivo de verificar, falsificar o establecer la validez de una hipótesis.
- Visualización: Es la representación visual de datos abstractos para reforzar la cognición humana.
- Teórico: En el que surgen nuevos conceptos a partir del conocimiento científico existente.
- Computacional: En el que descubrimos nuevos principios basándonos en algunos experimentos computacionales.
- Data-Driven: Basado en el descubrimiento de relaciones y patrones ocultos en los datos.

### 1.3. Campos relacionados <a class="anchor" id="t1_3"></a>

Los principales campos relacionados son: matemáticas analíticas (incluidas las estadísticas) y visualización.

**Cantidad**
- El estudio de la cantidad comienza con los números, primero los números naturalesm enteros, operaciones aritméticas básicas. A medida que se desarrolla más el sistema numérico, los números enteros se reconocen como un subconjunto de los números racionales. Estos, a su vez, están contenidos dentro de los números reales, que se utilizan para representar cantidades continuas. Los números reales se generalizan a números complejos.

**Estructura**
- Muchos conjuntos de objetos matemáticos exhiben una estructura interna. Las matemáticas exponen estas estructuras aplicando reglas (axiomas y operaciones) a los objetos. El álgebra es una poderosa herramienta para comprender las estructuras matemáticas. Combina el concepto de variables con la aritmética para resolver ecuaciones.

**Espacio**
- El estudio avanzado del espacio incluye geometría de dimensiones superiores, geometrías no euclidianas, geometría diferencial, topología, geometría fractal.


**Cambio**
- Comprender y describir el cambio es un tema común en la ciencia, y el cálculo se desarrolló como una poderosa herramienta para investigarlo. Las funciones son un concepto central que describe una cantidad cambiante. Muchos problemas conducen naturalmente a relaciones entre una cantidad y su tasa de cambio.

**Matemáticas Aplicadas**
- Las matemáticas aplicadas se ocupan de los métodos matemáticos que se utilizan normalmente en la ciencia, la ingeniería, los negocios y la industria. Por lo tanto, las "matemáticas aplicadas" son matemáticas con conocimientos especializados. En términos generales, este es el tipo de matemáticas que practican los científicos de datos.


Dado que los datos son omnipresentes, la ciencia de datos en sí también es un campo amplio que toca muchas otras disciplinas.
- Bases de datos: Una consideración critica es “cómo almacenar los datos”, de qué forma estructurarlos para poder procesarlos de una forma más eficiente
- Big Data: A menudo necesitamos almacenar y procesar grandes cantidades de datos con una estructura sencilla. Existen enfoques y herramientas para almacenar estos da tos de forma distribuida en clústeres y procesarlos de forma eficiente.
- Aprendizaje automático: Una forma de entender los datos es construyendo un modelo con el cuál seremos capaces de predecir una salida deseada.
- Inteligencia Artificial: Esté área también se basa en datos, y consiste en construir modelos de alta complejidad que imitan los procesos del pensamiento humano.
- Visualización: Estas visualizaciones nos permiten interpretar los datos de una manera más sencilla dándolos de sentido y permitiéndonos obtener conclusiones.

### 1.4. Cómo obtener datos <a class="anchor" id="t1_4"></a>

Existen multitud de maneras de obtener datos, se listan algunos de los más típicos:

Estructurados
- Internet of Things: Se incluyen datos de distintos sensores que proveen datos valiosos.
- Encuestas o sondeos: Los usuarios las completan después de una compra o visitar un sitio.
- Análisis de comportamiento: Nos puede ayudar por ejemplo saber como un usuario interactúa con el sitio web, y cuál es la principal razón de que lo abandone.

Sin estructurar
- Texto: Pueden ser una rica fuente de información, o puntuación general del sentimiento. 
- Imágenes o videos: Un video de una cámara de vigilancia puede ser usado para estimar el tráfico en una carretera.
- Registros o logs web: Se puede utilizar para entender que paginas de un sitio web son visitas con mayor frecuencia, y la duración de la visita.

Semi estructurar
- Redes sociales: Son grafos que nos permiten obtener datos sobre la personalidad de los usuarios y su eficiencia a la hora de difundir información. 
- Dinámicas de grupo: Se puede construir un grafo que relacione los usuarios que tienen una relación entre sí.


### 1.5. ¿Qué podemos hacer con los datos? <a class="anchor" id="t1_5"></a>

**Ingeniería de datos**

La ingeniería de datos consiste en adquirir, ingerir, transformar, almacenar y recuperar datos. La ingeniería de datos también incluye añadir metadatos a los datos.

*¿Cuál es el proceso típico de Ciencia de Datos?*

1.	Adquisición de datos
    - Los datos pueden provenir en variedad de formas y lugares, el primer paso es colectar los datos de una forma coherente para seguir con el resto de procesos.
2.	Almacenamiento de datos
    - Almacenar datos puede ser algo complejo y desafiante si hablamos de grandes volúmenes de datos. Tiene sentido anticipar la forma en que se consultarán en el futuro:
        -  Base de datos relacional: Almacena los datos en tablas utilizando SQL, donde las tablas están organizadas en grupos llamados esquemas.
        - Bases de datos NoSQL: No utiliza esquemas y nos permite almacenar datos más complejos como grafos y documentos JSON jerárquicos, pero no tienen la capacidad de consulta de SQL, ni reglas sobre cómo se estructuran los datos en las tablas.
        - Data Lake: Se utilizan para grandes colecciones de datos en bruto y sin estructurar, se utilizan Big Data almacenados y procesados por un clúster de servidores.
3.	Procesamiento o transformación de datos
    - Implica la conversión de datos de su forma original a una forma que puede ser usada para visualización/modelo de entrenamiento. Cuando procesamos datos sin estructurar podemos utilizar técnicas AI para extraer características y convertirlas en datos estructurados.
4.	Visualización
    - Para poder comprender mejor los datos necesitamos visualizarlos, muchas veces es necesario visualizándolos muchas veces y buscando algunas relaciones. También podemos utilizar técnicas estadísticas para probar una hipótesis o demostrar una correlación entre distintos datos.
5.	Entrenamiento de un modelo predictivo
    - Dado que el objetivo es poder tomar decisiones basadas en datos, es posible que queramos utilizar técnicas de aprendizaje automático para construir un modelo predictivo. Pudiendo así utilizarlo para hacer predicciones con nuevos conjuntos de datos similares.

### 1.6. Digitalización y transformación digital <a class="anchor" id="t1_6"></a>

En la última década las empresas empezaron a entender la importancia de los datos a la hora de tomar decisiones de negocio. Para aplicar la ciencia de datos a un negocio, uno primero debe colectar estos datos, traducir los procesos del negocio a una forma digital conocido como digitalización. Aplicando la ciencia de datos a estaos datos nos permite guiarnos y tomar decisiones que nos permitan aumentar significativamente la productividad o hacer una transformación digital.

La disponibilidad de datos administrativos/organizativos crea otras oportunidades. Los investigadores pueden vincular los datos recopilados de estudios científicos con, por ejemplo, datos médicos que se recopilaron con fines de atención médica; en otras palabras, los datos administrativos recopilados por razones que no se derivan directamente de la pregunta de interés pueden ser útiles en otros entornos. Dichos vínculos pueden ayudar a los científicos de datos a ampliar las posibilidades de sus análisis y verificar la calidad de sus datos.

### 1.7. El impacto de la ciencia de datos <a class="anchor" id="t1_6"></a>

Algunos ejemplos de cómo la ciencia de datos ha revolucionado varios aspectos diferentes de nuestro mundo: Béisbol, Salud y Robótica.

**Moneyball**

Se refiere al libro de Michael Lewis, publicado en 2003, sobre el equipo de béisbol Oakland Athletics y su director general Billy Beane, y posteriormente en 2011 se estrenó una película basada en el libro.

La premisa central es que la sabiduría acumulada por los entendidos del béisbol es subjetiva y a menudo errónea. El libro sostiene que la directiva de los Oakland A's aprovechó las ventajas de unos indicadores más analíticos del rendimiento de los jugadores para formar un equipo que pudiera competir con éxito contra los rivales más ricos de la Major League Baseball.

Rigurosos análisis estadísticos habían demostrado que el porcentaje de bases y el porcentaje de slugging son mejores indicadores del éxito ofensivo, y los A's se convencieron de que estas cualidades eran más baratas de obtener en el mercado que otras más valoradas históricamente, como la velocidad y el contacto. Estas observaciones a menudo iban en contra de la sabiduría convencional del béisbol y de las creencias de muchos ojeadores y ejecutivos.

Los tradicionalistas del béisbol, en particular algunos ojeadores y miembros de los medios de comunicación, condenan la revolución sabermétrica y han menospreciado Moneyball por hacer hincapié en los conceptos sabermétricos por encima de los métodos más tradicionales de evaluación de jugadores. Sin embargo, el impacto de Moneyball en las oficinas de las grandes ligas es innegable.

**23 and Me**

Es una empresa privada de genómica personal y biotecnología con sede en Mountain View, California, que ofrece pruebas genéticas rápidas.

23andMe comenzó a ofrecer servicios de pruebas de ADN en noviembre de 2007, cuyos resultados se publican en línea y permiten al suscriptor ver una evaluación de rasgos hereditarios, genealogía y posibles factores de riesgo congénitos.

La organización también ofrece pruebas para determinadas iniciativas de investigación, proporciona conjuntos de datos confidenciales con el objetivo de establecer asociaciones genéticas con enfermedades y trastornos específicos. El cofundador de Google, Sergey Brin (cuya madre padece la enfermedad de Parkinson y está casado con la cofundadora de 23andMe, Anne Wojcicki), sufragó el coste de la Iniciativa Genética contra la Enfermedad de Parkinson de la empresa para ofrecer pruebas gratuitas a las personas que padecen esta dolencia. Un análisis de los resultados de la investigación sobre la enfermedad de Parkinson que comparaba a 23andMe con una iniciativa de los Institutos Nacionales de Salud (NIH) sugería que el uso por parte de la empresa de grandes cantidades de potencia computacional y conjuntos de datos podría ofrecer resultados comparables, aunque en mucho menos tiempo.


**Google's Driverless Car**

El sistema combina información obtenida de Google Street View con software de inteligencia artificial que combina datos de cámaras de vídeo situadas en el interior del coche, un sensor LIDAR en la parte superior del vehículo, sensores de radar en la parte delantera del vehículo y un sensor de posición fijado a una de las ruedas traseras que ayuda a localizar la posición del coche en el mapa.

Aunque Google no tenía planes inmediatos de desarrollar comercialmente el sistema, la empresa espera desarrollar un negocio que comercialice el sistema y los datos que lo respaldan a los fabricantes de automóviles. Un abogado del Departamento de Vehículos a Motor de California expresó su preocupación por el hecho de que "la tecnología va por delante de la ley en muchos ámbitos", citando leyes estatales que "todas presuponen un humano conduce".

Sólo 5 años después, el coche sin conductor de Google había recorrido cientos de miles de kilómetros a una velocidad límite de 55 a 65 millas por hora. ¿Avanzó tanto la disciplina de la inteligencia artificial en 5 años? No. La diferencia fue la ciencia de datos. Los científicos de datos de Google crearon un modelo de malla tridimensional de cada calle por la que iba a circular el coche sin conductor. En tiempo real, los algoritmos de la ciencia de datos del coche comparaban las observaciones reales con el modelo (incluidas las rayas blancas de la carretera) y hacían las correcciones pertinentes.

## 2. Ética de la ciencia de datos <a class="anchor" id="t2"></a>

### 2.1. Introducción a la ética <a class="anchor" id="t2_1"></a>

La ética abarca los valores que compartimos y los aspectos morales que gobiernan este comportamiento.
Ética de datos es un nuevo campo de la ética que estudia y evalúa problemas morales relacionados con los datos, algoritmos, y procesos. Es necesario aplicar unas consideraciones morales, procesos de una investigación constante de problemas éticos del mundo que nos rodea, para que se alinea con nuestros valores morales.
La cultura ética consiste en hacer operativa la ética aplicada para asegurarnos de que nuestros principios y prácticas éticos se adoptan de forma coherente y escalable en toda la organización.


### 2.2. Principios éticos <a class="anchor" id="t2_2"></a>

Toda estrategia de ética de los datos empieza por definir unos principios éticos compartidos que describen un comportamiento aceptable, podemos identificar 6 principios éticos:
- Accountability: Responsabiliza a los profesionales de sus operaciones con datos e IA y del cumplimiento de estos principios éticos.
- Transparency: Garantiza que los datos y las acciones de IA sean comprensibles (interpretables) para los usuarios, explicando el qué y el porqué de las decisiones.
- Fairness: Se centra en garantizar que la IA trate a todas las personas de forma justa, abordando cualquier sesgo sociotécnico sistémico o implícito en los datos y sistemas.
- Reliability & Safety: Garantiza que la IA se comporte de forma coherente con los valores definidos, minimizando posibles daños o consecuencias imprevistas.
- Privacity & Safety: Se trata de comprender el linaje de los datos y de proporcionar a los usuarios protección de la privacidad y otras protecciones relacionadas.
- Inclusivness: Se trata de diseñar soluciones de IA con intención, adaptándolas para satisfacer una amplia gama de necesidades y capacidades humanas.


### 2.3. Desafíos éticos <a class="anchor" id="t2_3"></a>

Una vez definidos los principios éticos, el siguiente paso es evaluar nuestros datos y acciones para comprobar si se ajustan a nuestros valores. A continuación, se detallan algunos de los desafíos: 


1.Data Ownership: La recogida de datos suele implicar datos personales que pueden identificar a los interesados. La propiedad de los datos se refiere al control y los derechos de los usuarios en relación con la creación, el tratamiento y la difusión de los datos.
<ul>
    <li> ¿A quién pertenecen los datos? (usuario u organización) </li>
    <li> ¿Qué derechos tienen los interesados? (por ejemplo: acceso, supresión, portabilidad) </li>
    <li> ¿Qué derechos tienen las organizaciones? (rectificar reseñas de usuarios) </li>
</ul>
2.Informed Consent: El consentimiento informado define el acto por el que los usuarios aceptan una acción (como la recogida de datos) con pleno conocimiento de los hechos relevantes, incluidos el propósito, los riesgos potenciales y las alternativas.
<ul>
    <li>¿Ha dado el usuario (sujeto de los datos) permiso para la recogida y uso de los datos?</li>
    <li>¿Comprendió el usuario la finalidad de la recogida de datos?</li>
    <li>¿Comprendió el usuario los riesgos potenciales de su participación?</li>

## 3. Definición de los datos <a class="anchor" id="t3"></a>
![title](../img/Divider.png)

### 3.1. ¿Cómo son descritos los datos? <a class="anchor" id="t3_1"></a>

Los datos son hechos, información, observaciones y mediciones que usados para hacer descubrimientos y sobre los que nos apoyamos para tomar decisiones. Un punto de datos es una unidad de datos dentro de un conjunto de datos, que es una colección de puntos de datos. Los conjuntos de datos pueden tener distintos formatos y estructuras, y suelen basarse en su origen, es decir, en la procedencia de los datos.
- Raw data: Son datos que vienen de un origen donde su estado inicial aún sin organizar o analizar. Para poder entender estos datos debemos estructurarlos y organizarlos.
- Quantitative Data: Son observaciones numéricas dentro de un conjunto de datos y normalmente pueden analizarse, medirse y utilizarse matemáticamente.
- Qualitative Data: Los datos cualitativos o categóricos, son datos que no pueden medirse objetivamente. Por lo general, se trata de diversos formatos de datos subjetivos que captan la calidad de algo, como un producto o un proceso. A veces, los datos cualitativos son numéricos y no se suelen utilizar matemáticamente, como los números de teléfono o las marcas de tiempo.
- Structured Data: Los datos estructurados son datos organizados en filas y columnas. Las columnas suelen tener un conjunto específico de reglas o restricciones sobre los valores, para garantizar que los valores representan con precisión. Una de las ventajas de los datos estructurados es que pueden organizarse de tal forma que pueden relacionarse con otros datos estructurados.
- Unstructured Data: Por lo general, no pueden clasificarse en filas o columnas y no contienen un formato o conjunto de reglas a seguir. Esto puede hacer que analizar o investigar este tipo de datos lleve más tiempo y disponer de datos completos. 
- Semi-Structured: Son una combinación de datos estructurados y no estructurados. No suelen ajustarse a un formato de filas y columnas, pero se organizan de una manera que se considera estructurada y pueden seguir un formato fijo o un conjunto de reglas. Los metadatos son indicadores que ayudan a decidir cómo se organizan y almacenan los datos y tendrán varios nombres, en función del tipo de datos. Algunos nombres comunes para los metadatos son etiquetas, elementos, entidades y atributos.


### 3.2. Alcance de los datos <a class="anchor" id="t3_2"></a>

Un paso inicial importante en el ciclo de vida de los datos es expresar la pregunta de interés en el contexto del área temática y considerar la conexión entre la pregunta y los datos recopilados para responder esa pregunta.

Como parte de hacer la conexión entre el proceso de recolección de datos y el tema de investigación, identificamos la población, los medios para acceder a la población, los instrumentos de medición y los protocolos adicionales utilizados en el proceso de recolección.

La población objetivo consiste en la colección de elementos que, en última instancia, tiene la intención de describir y sacar conclusiones. Por elemento entendemos aquellos individuos que componen nuestra población.

El marco de acceso es la colección de elementos que son accesibles para su medición y observación. Estas son las unidades por las que se puede acceder a la población objetivo. Idealmente, el marco y la población están perfectamente alineados; lo que significa que consisten en exactamente los mismos elementos.

Los contenidos del marco de acceso, en comparación con la población objetivo, y el método utilizado para seleccionar las unidades del marco para estar en la muestra son factores importantes para determinar si los datos pueden considerarse representativos de la población objetivo o no.

En muchos experimentos y estudios, no tenemos la capacidad de incluir todas las unidades de población en el marco. A menudo ocurre que el marco de acceso consiste en voluntarios que están dispuestos a unirse al estudio/experimento.

Por ejemplo, las encuestas preelectorales se llevan a cabo regularmente durante la campaña presidencial, y cuanto más se acerque el día de las elecciones, esperamos que las encuestas mejoren en la predicción del resultado, a medida que se estabilicen las preferencias. Las encuestas también se realizan típicamente en todo el estado y luego se combinan para hacer predicciones para el ganador general. 

El marco de acceso consiste en aquellos que se consideran votantes probables (esto generalmente se basa en su historial de votación anterior, pero se pueden usar otros factores para determinar esto), y dado que se contacta a las personas por teléfono, el marco de acceso está limitado a quienes tienen teléfono fijo o móvil. La muestra está formada por aquellas personas del marco que se eligen de acuerdo con un esquema de marcación aleatoria.

### 3.3. Protocolos e instrumentos <a class="anchor" id="t3_3"></a>

Cuando consideramos el alcance de los datos, también consideramos el instrumento que se utiliza para tomar las medidas y el procedimiento para tomar las medidas, al que llamamos protocolo.

Los protocolos también son críticos en los experimentos. Idealmente, se controla cualquier factor que pueda influir en el resultado del experimento. Por ejemplo, la temperatura, la hora del día, la confidencialidad de un registro médico e incluso el orden en que se toman las medidas deben ser coherentes para descartar los posibles efectos de estos factores que interfieren.

Con los rastros digitales, los algoritmos utilizados para respaldar la actividad en línea son dinámicos y se rediseñan continuamente. Por ejemplo, los algoritmos de búsqueda de Google se modifican continuamente para mejorar el servicio al usuario y los ingresos publicitarios. Los cambios en los algoritmos de búsqueda pueden afectar los datos generados a partir de las búsquedas, lo que a su vez afecta los sistemas creados a partir de estos datos, como el sistema de seguimiento Google Flu Trend. Este entorno cambiante puede hacer que sea insostenible mantener los protocolos de recopilación de datos y dificultar la replicación de los hallazgos.

Muchos proyectos implican vincular datos de múltiples fuentes. Cada fuente debe ser examinada y cualquier diferencia entre las fuentes debe ser considerada. Además, los algoritmos de coincidencia utilizados para combinar datos de múltiples fuentes deben entenderse claramente para que se puedan comparar las poblaciones y los marcos de las fuentes.

### 3.4. Exactitud <a class="anchor" id="t3_4"></a>

En la mayoría de los entornos, necesitamos cuantificar la precisión de nuestras mediciones para generalizar nuestros hallazgos a lo que no se observa. Por ejemplo, a menudo usamos la muestra para estimar un valor promedio para una población, inferir el valor de una incógnita científica a partir de mediciones o predecir el comportamiento de un nuevo individuo. En cada una de estas configuraciones, también queremos un grado cuantificable de precisión. Queremos saber qué tan cerca de la verdad están nuestras estimaciones, inferencias y predicciones.

El sesgo viene en muchas formas. Se describen algunos tipos clásicos:
- El sesgo de cobertura ocurre cuando el marco de acceso no incluye a todos en la población objetivo. 
- El sesgo de selección surge cuando el mecanismo utilizado para elegir las unidades de la muestra tiende a seleccionar ciertas unidades con más frecuencia de la que debería.
- El sesgo de no respuesta.
- El sesgo de medición ocurre cuando un instrumento falla sistemáticamente el objetivo en una medición.

Los protocolos son clave para reducir estas fuentes de sesgo. Los mecanismos de azar para seleccionar una muestra del marco o para asignar unidades a las condiciones experimentales pueden eliminar el sesgo de selección. Un protocolo de seguimiento de la falta de respuesta para alentar la participación puede reducir el sesgo de falta de respuesta. Una encuesta piloto puede mejorar la redacción de las preguntas y así reducir el sesgo de medición.

## 4. Introducción a Estadística y Probabilidad <a class="anchor" id="t4"></a>

### 4.1. Definición de Estadística <a class="anchor" id="t4_1"></a>

La estadística se ocupa de la recogida, el análisis, la interpretación y la presentación de datos, para poder hacer comparaciones y sacar conclusiones para elfuturo.


Así pues, lo primero de todo es el ESTUDIO ESTADÍSTICO, que consta de las siguientes fases:

- Recogida de datos.
- Organización y representación de datos.
- Análisis de datos.
- Obtenciónde conclusiones.

Debemos empezar con la correcta definición de una serie de conceptos estadísticos.
- **Población**: Una población (o UNIVERSO) es el conjunto de todos los elementos a los que se somete a un estudio estadístico.
- **Individuo**: Un individuo o unidad estadística es cada uno de los elementos que componen la población.Sería codo uno de /as personas de nuestro ciudad
- **Muestra**: Una muestra es un conjunto representativo de la población de referencia, el número deindividuos de una muestra es siempre menor que el dela población. El problema puede llegar a ser el elegir una muestra que de verdad sea representativa. 
- **Valor**: Un valor es cada uno de los distintos resultados que se pueden obtener en un estudio estadístico.
- **Dato**: Un dato es cada uno de los valores que se ha obtenido al realizar un estudio estadístico.

### 4.2. Datos, muestreo y variación de los datos y el muestreo <a class="anchor" id="t4_2"></a>

Variable estadística: cada una de las características consideradas con el propósito de describir a cada
individuo de la muestra.


Las variables cualitativos son el resultado de categorizar o describir atributos de una población. Los datos cualitativos también suelen denominarse datos categóricos.
<ul>
    <li>Variable cualitativa nominal: Una variable cualitativa nominal presenta modalidades no numéricas que noadmiten un criferio de orden. </li>
    <li>Variable cualitativa ordinal: Una variable cualitativa ordinal presenta modalidades no numéricas. en las que existe un orden. es decir, unas modalidades tienen más peso que otras.</li>
    <li>Variable cualitativa dicotómica: Una voriable cualitativa ordinal presenta modalidades no numéricos, en las que sólo hay dos modalidades  diferentes. </li>
</ul>

Las variables cuantitativos son siempre números. Los datos cuantitativos son el resultado de contar o medir atributos de una población.
<ul>
    <li>Variable discreta: Una variable discreto es aquella que toma valores aislados, es decir, no admite valores intermedios entre dos volores específicos.</li>
    <li>Variable continua: Una voriable continua es aquella que pueden tomar valores comprendidos entre dos números (decimales).</li>
</ul>



**Muestreo**
Reunir información sobre toda una población suele costar demasiado o ser prácticamente imposible. En su lugar, se utiliza una muestra de la población. Una muestra debe tener las mismas características que la población a la que representa. La mayoría de los estadísticos utilizan diversos métodos de muestreo aleatorio para intentar alcanzar este objetivo.

_Métodos conocidos de muestreo aleatorio_

- El método más fácil de describir se denomina muestra aleatoria simple. En una muestra aleatoria simple, cada grupo tiene la misma probabilidad de ser seleccionado. En otras palabras, cada muestra del mismo tamaño tiene la misma probabilidad de ser seleccionada.
- Para elegir una muestra estratificada, se divide la población en grupos denominados estratos y, a continuación, se selecciona la muestra escogiendo el mismo número de valores de cada estrato hasta alcanzar el tamaño de muestra deseado.
- Para elegir una muestra de conglomerados, se divide la población en conglomerados (grupos) y, a continuación, seleccione al azar algunos de los conglomerados.
- Para elegir una muestra sistemática, se selecciona una muestra sistemática consiste en enumerar los miembros de la población y elegir cada enésima entrada a partir de un punto de partida aleatorio.






**Evaluación crítica** 

Al analizar los datos, es importante tener en cuenta los errores de muestreo y los errores ajenos al muestreo.
En realidad, una muestra nunca será exactamente representativa de la población, por lo que siempre habrá algún error de muestreo. Por regla general, cuanto mayor sea la muestra, menor será el error de muestreo.

En estadística, se crea un sesgo de muestreo cuando se recoge una muestra de una población y algunos miembros de la población no tienen las mismas probabilidades de ser elegidos que otros. Recuerde que cada miembro de la población debe tener la misma probabilidad de ser elegido. Cuando se produce un sesgo en el muestreo, pueden extraerse conclusiones incorrectas sobre la población estudiada.


Debemos evaluar críticamente los estudios estadísticos que leemos y analizarlos antes de aceptar sus resultados. Entre los problemas habituales que hay que tener en cuenta se incluyen los siguientes:
- Problemas con las muestras: Una muestra debe ser representativa de la población.
- Muestras autoseleccionadas: Las respuestas sólo de personas que deciden responder.
- Cuestiones relacionadas con el tamaño de la muestra: Las muestras demasiado pequeñas pueden resultar poco fiables. Las muestras más grandes son mejores, si es posible.
- Influencia indebida: Eecopilación de datos o formulación de preguntas que influyen en la respuesta.
- Falta de respuesta o negativa del sujeto a participar.
- Causalidad: Una relación entre dos variables no significa que una provoque la otra.

### 4.3. Distribución de frecuencia <a class="anchor" id="t4_3"></a>

#### 4.3.1. Distribución de frecuencia <a class="anchor" id="t4_3_1"></a>

La distribución de frecuencias o tabla de frecuencias es una ordenación en forma de tabla de los datos estadísticos,asignando a cada dato su frecuencia correspondiente.

Tipos de frecuencia
- **Frecuencia absoluta**: La frecuencia absoluta es el número de veces que aparece un determinado valor enun estudio estadístico. La suma de las frecuencias absolutas es igual al número total de datos, representado por N.
- **Frecuencia relativa**: La frecuencia relativa es el cociente entre la frecuencia absoluta de un determinado valor ni y el número total de datos N.
- Frecuencia absoluta acumulada: La frecuencia acumulada es la suma de las frecuencias absolutas de todos los valores inferiores o iguales al valor considerado.
- Frecuencia relativa acumulada: La frecuencia relativa acumulada es elcociente entrela frecuencia acumulada de un determinado valor y el número total de datos.

##### 4.5.1.1. Medidas de centralización <a class="anchor" id="t4_5_1_1"></a>

Nos indican en torno a qué va lor (centro) se distribuyen los datos. Nos da la información de hacia dónde tiende todos los datos para obtener un valor lo más próximo ala realidad posible.


**Media aritmética**

- La media es el promedio de una distribución. Suele ser el valor que mós se repite y se acepta en muchas hipótesis como el valor verdadero frente a errores.
- Solo es aplicable a datos cuantifativos.
- La media es muy sensible a las puntuaciones extremas, pudiendo no ser tan representativa conla realidad ya que con un solo valor erróneo debido una mal medición, cambiaría toda la media.


```python
media = df['age'].mean()
print(f'La media es de {media:4.1f} años')
```

**Mediana**

- La mediano es la puntuación de lo escala que separa la mitad superior de lo distribución y la inferior, es decir, divide la serie de datos en dos partes iguales con el mismo número de datos a coda lado.
- Así pues,es el valor que ocupo el lugar central de todos los datas cuando éstos están ordenados de menor o mayor.
- La mediana se puede hallar sólo para variables cuantitativas.


```python
mediana = df['age'].median()
print(f'La mediana de la edad es {int(mediana)} años')
```

**Moda**
- La moda es el valor que más se repite en la distribución (tiene mayor frecuencia absoluta), por lo que podemos entender que las medidas tienden a ese valor, aunque no es tan representativo dependiendo del suceso.


```python
moda = df['blood_type'].mode()[0]
print(f'El tipo sanguíneo más común es {moda}')

# Representacion grafica de la moda
plt.figure(figsize=(10,6))
plt.hist(df['blood_type'], bins=20)
plt.tight_layout()
plt.show()
```

##### 4.5.1.2. Medidas de posición <a class="anchor" id="t4_5_1_2"></a>

Las medidas de posición dividen un conjunto de datos en grupos con el mismo número de individuos. Para calcular las medidas de posición es necesario que los datos estén ordenados de menor a mayor. 

**Cuartiles**
- Los cuartiles dividen la serie de datos en cuatro partes iguales,con el mismo número de datos encada parte.
- Los cuartiles son los tres valores de la variable que dividen a un conjunto de datos ordenados en cuatro partes iguales y se nombran Q1,Q2 y Q3 y determinan los valores correspondientes al 25%, al 50% y al 75% de los datos.
- Q2 se corresponde con la mediana 50%.


```python
# Five number summary
minimum = df['age'].min()
first_quartile = df['age'].quantile(0.25)
median = df['age'].median()
third_quartile = df['age'].quantile(0.75)
maximum = df['age'].max()
print('Five number summary: ', 'Mínimo:', minimum, '- Q1:', first_quartile, '- Q2 o mediana:', median, '- Q3:', third_quartile, '- Máximo:', maximum)
```

**Percentiles**
- Los percentiles dividen la serie de datos en cien partes iguales.
- Los percentiles son los 99 valores que dividen la serie de datos en 100 partes iguales, por lo tanto dan los valores correspondientes al 1%.2%.3%....,98%,99% de los datos.
- El valor del percentil P50 coincide con la mediana.


```python
# Percentiles
tenth_percentile = df['age'].quantile(0.1)  # 10th percentile
ninetieth_percentile = df['age'].quantile(0.9)  # 90th percentile
print(f'El percentil 10 es de {tenth_percentile:4.1f} años; El percentil 90 es de {ninetieth_percentile:4.1f} años')

# Funcion auxiliar de Seaborn
def plot_errorbars(arg, **kws):
    np.random.seed(sum(map(ord, "error_bars")))
    x = df['age']
    f, axs = plt.subplots(2, figsize=(7, 2), sharex=True, layout="tight")
    sns.pointplot(x=x, errorbar=arg, **kws, capsize=.3, ax=axs[0])
    sns.stripplot(x=x, jitter=.3, ax=axs[1])

# Representa donde se agrupa el 75% de los datos y en que rango se agrupan mejor
plot_errorbars(("pi", 75))
```

Propiedades de la varianza

- La varianza no viene expresada en las mismas unidades que los datos, ya que las desviaciones están elevadas al cuadrado. Es un índice adimensional.


```python
varianza = df['age'].var()
print('La varianza es', varianza)
```

**Desviación típica**
- Es la raíz cuadrada d ela varianza
- Representa el error típico cometido cuando realizamos un estudio en busca del valor cerdadero de algún parámetro.

Propiedades de la desviación típica
- Será siempre un **valor positivo o cero**, en caso de que las puntuaciones sean iguales.

**Coeficiente de Asimetría de Fisher**

Esta media nos permite identificar si los datos se distribuyen de forma uniforme alrededor del punto central. 

La asimetría representa tres estados diferentes, cada uno de los cuales define de forma concisa como están distribuidos los datos respecto al eje de asimetría:

- Coeficiente de asimetría > 0 asimetría positiva
- Coeficiente de asimetría = 0 simetrica
- Coeficiente de asimetría < 0 asimetría negativa

<center><img src="../img/Asimetria.jpeg"/></center>

```python
skewness = skew(df['age'])
print("Coeficiente de Asimetría (Fisher):", skewness, ' > 0 nos indica que se trata de una asimetría positiva')

print('Mediana', df['age'].median())
print('Media', df['age'].mean())
print('También nos podemos fijar en la relación entre la mediana que se encuentra desplazada a la izquierda de la media.')
print('Así mismo podemos observar en la distribución de densidad que existe una asimetría positiva de los datos')

# Funcion para mostrar la distribucion de densidad
def plot_density_distribution(data, xlabel, ylabel, title):
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data, color='blue', fill=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

plot_density_distribution(df['age'], 'Edad', 'Densidad', 'Distribución de Densidad')
```

- Subjetiva (bayesiana): Grado de certeza que se posee sobre un suceso. Es personal.



#### 4.6.1. Introducción a la Probabilidad <a class="anchor" id="t4_6_1"></a>

**Experimentos y sucesos aleatorios**

Experimento aleatorio es aquel en el que se conoce el conjunto de resultados posibles, pero se desconoce el resultado para cada experimento.

La probabilidad es una medida asociada a la certeza de los resultados de una determinada actividad. Cuando la actividad es una operación planificada que se lleva a cabo en condiciones controladas, se denomina experimento. Si el resultado no está predeterminado, se dice que el experimento es fortuito. Cada vez que se intenta el experimento se denomina ensayo.

El conjunto de todos los resultados posibles (sucesos elementales), cuando se realiza un experimento aleatorio, se llama espacio
muestral (E). Se llama suceso aleatorio a un subconjunto del espacio muestral.

- Suceso elemental cada uno de los resultados elementales del experimento aleatorio.
- Suceso seguro a aquel que se verifica siempre, o sea E y suceso imposible al que nunca se verifica, el conjunto vacío.
- Suceso contrario (complementario) de un suceso A, A’, al formado por los elementos que no están en A.
- Suceso unión de A y B, AUB, al formado por los resultados experimentales que están en A o en B (incluyendo los que
están en ambos).
- Suceso intersección de A y B, A∩B o simplemente AB, al formado por los resultados experimentales que están simultáneamente en A y B.

Algunos ejemplos de sucesos o experimentos son los siguientes:
- Lanzar una moneda al aire,
- Hacer girar una ruleta,
- Tirar un par de dados.

Un suceso es cualquier combinación de resultados. Es un subconjunto del espacio muestral, por lo que se suelen utilizar letras mayúsculas como A y B para representar los sucesos. Por ejemplo, si el experimento consiste en lanzar tres monedas justas, el suceso A podría ser obtener como máximo una cara.

**Definición axiomática de probabilidad (Kolmogorov)**

Dado un espacio muestral E, diremos que P es una función de probabilidad sobre E si se verifican los siguientes axiomas:
- Axioma 1. La probabilidad es una función definida sobre E y que sólo toma valores positivos comprendidos entre 0 y 1.
    - Dado A pertenece a E, se cumple que 0≤P(A)≤1.P(A)=0
- Axioma 2. La probabilidad del suceso seguro es 1.
    - P(A)=1
- Axioma 3. La probabilidad de la unión numerable de sucesos disjuntos (incompatibles) es la suma de sus probabilidades.

#### 4.7.1. Variables aleatorias <a class="anchor" id="t4_7_1"></a>

- Variable aleatoria: A cada suceso elemental, e, del espacio muestra E le atribuye un único número real.
- Experimento: Proceso de obtener un dato dadas unas condiciones de experimentación. La obtención de un nuevo dato, con las mismas condiciones, es una repetición del experimento.
- Experimento aleatorio: Experimento que al repetirlo no siempre se obtiene el mismo valor.
- Variable aleatoria: Es el resultado de un experimento aleatorio. Es la variable respuesta de un experimento aleatoria
- Valores observados: Los valores observados se les denomina realizaciones de la variable aleatoria.

Clasificación de las variables aleatorias:

- Continuas: Pueden tomar un número infinito no numerable de valores, significa que los valores se obtienen haciendo una medición.

#### 4.7.2. Distribuciones de Probabilidad <a class="anchor" id="t4_7_2"></a>

Una distribución de frecuencias describe el número de observaciones para cada valor posible de una variable.

Una **distribución de probabilidad** es una distribución de frecuencias idealizada. Describe la probabilidad de diferentes valores posibles de una variable que viene determinado por su probabilidad de aparición. La probabilidad es un número entre 0 y 1 que indica la probabilidad de que ocurra algo.

Las muestras infinitamente grandes son imposibles en la vida real, por lo que las distribuciones de probabilidad son teóricas. Son versiones idealizadas de las distribuciones de frecuencia que pretenden describir la población de la que se extrajo la muestra.

Para representar estas distribuciones podemos usar:
- Funciones de densidad de probabilidad (PDF): Una función de densidad de probabilidad nos indica la probabilidad de que una variable aleatoria **discreta** tome un determinado valor. P(x = 1) : 1/6
    - Cada probabilidad está comprendida entre cero y uno, ambos inclusive.
    - La suma de las probabilidades es uno.
- Funciones de distribución de masa (CDF): Una función de distribución acumulativa nos indica la probabilidad de que una variable aleatoria **discreta o continua** tome un valor menor o igual que x. P(x ≤ 1) : 1/6

##### **4.7.2.1. Distribución Binomial** <a class="anchor" id="t4_7_2_1"></a>

Hay tres características de un experimento binomial:
- Hay un número fijo de ensayos p repeticiones del experimento, n indica el número de ensayos.
- Sólo hay dos resultados posibles, denominados éxito y fracaso, para cada ensayo. El resultado que estamos midiendo se define como éxito, mientras que el otro resultado se define como fracaso. La letra p indica la probabilidad de éxito en un ensayo y q la probabilidad de fracaso en un ensayo. p + q = 1.
- Los n ensayos son independientes y se repiten en las mismas condiciones. Como los n ensayos son independientes, el resultado de un ensayo no ayuda a predecir el resultado de otro ensayo. Otra forma de decir esto es que para cada ensayo individual, la probabilidad, p, de éxito y la probabilidad, q, de fracaso siguen siendo las mismas.

Simular lanzar una moneda 10 veces, registrar el número de caras y repetirlo 300 veces con NumPy:

```python
# We want to display the PMF on the plot from 0 to 10
x = np.linspace(0, 10, 91)
# Probability mass function
# (for 10 trials with a probability of 0.5 for both outcomes)
pmf = st.binom.pmf(x, 10, 0.5)

# Set the seed so that we get the same random numbers each time this code runs
np.random.seed(20230810)
# Record the number of heads you get when you flip a fair coin 10 times, and
# do this 300 times
number_heads = []
for _ in range(300):
    # Flip a fair coin 10 times
    results = np.random.binomial(n=1, p=0.5, size=10)
    # Record the number of heads
    number_heads.append(results.sum())

    # Create plot
ax = plt.axes()
title = """Probability Mass Function (PMF)
of a Binomial Distribution"""
ax.set_title(title)
ax.set_ylabel('Probability')
ax.set_xlabel('Potential Number of Heads')
# Plot the simulated results
label = '300 Sets of 10 Coin Flips'
ax.hist(
    number_heads, np.arange(11), align='left', density=True,
    label=label, color='gray', alpha=0.5, rwidth=0.15
)
# Plot the PMF
ax.scatter(x, pmf, c='k', lw=0.1, label='PMF')
# Remove the upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# Adjust axes
plt.xlim([0, 10])
plt.ylim([0, 0.3])
# Finish
ax.legend()
plt.show()
```

En el gráfico parece que la probabilidad de obtener 0 caras o 10 caras es cero. Esto no es del todo correcto, las probabilidades exactas son:

```python
print(
    f'La probabilidad de obtener 0 caras en 10 tiradas es {pmf[0]:.7f}\n' +
    f'La probabilidad de obtener 10 caras en 10 tiradas, también es {pmf[-1]:.7f}'
)
```

Las probabilidades son las mismas porque la probabilidad de obtener 10 caras es la misma que la de obtener 0 colas, y la probabilidad de obtener 0 caras es la misma que la de obtener 10 colas. Al lanzar las monedas se obtiene una distribución simétrica de los resultados porque cada resultado (cara o cruz) tiene la misma probabilidad.

Dada esta pequeña probabilidad de obtener 0 caras (igual a 1/1024
), incluso en 300 repeticiones de 10 lanzamientos la probabilidad de que no ocurra es bastante alta:

```python
pmf = st.binom.pmf(0, 300, 1 / 1024)
print(f'La probabilidad de nunca obtener 0 caras en 300 repeticiones de 10 tiradas: {pmf:.3f} ({pmf:.1%})')
```

```python
# We want to display the CDF on the plot from 0 to 10
x = np.linspace(0, 10, 1001)
# Cumulative distribution function
# (for 10 trials with a probability of 0.5 for both outcomes)
cdf = st.binom.cdf(x, 10, 0.5)

# Cumulative sum of random numbers
number_heads.sort()
auc = [np.trapz([0.1] * i, number_heads[:i]) for i in range(300)]

# Create plot
ax = plt.axes()
title = """Cumulative Distribution Function (CDF)
of a Binomial Distribution"""
ax.set_title(title)
ax.set_ylabel('Cumulative Probability')
ax.set_xlabel('Potential Number of Heads')
# Plot the CDF
ax.plot(x, cdf, c='k', lw=2, label='CDF')
# Plot the cumulative sum of the simulated results
label = '300 Sets of 10 Coin Flips'
ax.hist(
    number_heads, bins=9, density=True, cumulative=True,
    histtype='step', label=label
)
# Adjust axes
plt.xlim([0, 10])
plt.ylim([0, 1])
# Finish
ax.legend(frameon=False)
plt.show()
```

Cuando el éxito y el fracaso son igualmente probables y el número de ensayo es un número grande la distribución binomial es una distribución normal.

##### **4.7.2.2. Distribución Normal** <a class="anchor" id="t4_7_2_2"></a>

Representa una distribución de puntuaciones estandarizadas, denominadas puntuaciones Z, en contraposición a las puntuaciones brutas (los valores reales de los datos). Una puntuación Z indica el número de desviaciones estándar que una puntuación se sitúa por encima o por debajo de la media. Las puntuaciones Z permiten comparar puntuaciones de conjuntos de datos diferentes, con medias y desviaciones típicas distintas.

Puntuaciones Z
Si X es una variable aleatoria distribuida normalmente y X ~ N(μ, σ), entonces la puntuación z es Z=x-μ/σ

Indica cuántas desviaciones típicas tiene el valor x por encima, a la derecha, o por debajo, a la izquierda, de la media, μ. Los valores de x que son mayores que la media tienen puntuaciones z positivas, y los valores de x que son menores que la media tienen puntuaciones z negativas. Si x es igual a la media, entonces x tiene una puntuación z de cero.


La probabilidad de que una medición tenga un valor determinado depende de cuál sea ese valor: la probabilidad de que se acerque al valor medio o esperado es bastante alta, mientras que la probabilidad de que se aleje del valor esperado es bastante baja. Por lo tanto, la probabilidad se distribuye (difiere) en el rango de valores que podría tener la medida.

Esto puede verse en la "función de densidad de probabilidad" (FDP), la función de las probabilidades relativas de que una medición tenga un valor determinado. Es más alta en el centro y más baja en los extremos (colas) porque es más probable que una medición tenga un valor cercano a la media.

```python
# Create plot
ax = plt.axes()
# Probability density function (PDF)
x = np.linspace(-3, 3, 1000)
ax.plot(x, st.norm.pdf(x), 'k')
# Mean
ax.axvline(0, 0, st.norm.pdf(0) / 0.45, c='k', ls='--', label='Mean')
# Axes
ax.set_ylim([0, 0.45])
ax.set_xlim([-3, 3])
# Format
ax.set_title('Probability Density Function (PDF) of\nthe Standard Normal Distribution')
ax.set_ylabel('Relative Likelihood')
ax.set_xlabel('Distance of the Value of a Random Variable from the Mean')
ax.legend(frameon=False)
plt.show()
```

Por tanto, tiene más sentido hablar de la probabilidad de que una variable aleatoria tenga un valor de x

La probabilidad de que una variable aleatoria tenga un valor dentro de un intervalo corresponde al área bajo la PDF de ese intervalo (porque, en esencia, es la suma de todas las probabilidades de cada uno de los infinitos valores individuales posibles).
```python
# Plot the sampling distribution
ax = plt.axes()
# Histogram
n, bins, patches = ax.hist(sample_means, color='gray', bins=35)
# Normal distribution
x_lim = ax.get_xlim()
x = np.linspace(x_lim[0], x_lim[1], 1000)
loc = np.mean(sample_means)
scale_x = np.std(sample_means)
scale_y = 1000 * np.mean(np.diff(bins))
y = st.norm.pdf(x, loc=loc, scale=scale_x) * scale_y
ax.plot(x, y, c='k', label='Normal distribution')
# Set x-axis limits
ax.set_xlim(x_lim)
# Format
ax.set_title('The Means of 1000 Samples of 10 Dice Rolls')
ax.set_ylabel('Frequency')
ax.set_xlabel('Mean Value Seen on Dice')
ax.legend(frameon=False)
plt.show()
```

##### **4.7.2.3. Distribución Uniforme** <a class="anchor" id="t4_7_2_3"></a>

Si alguien quiere generar un "número aleatorio", normalmente quiere decir que quiere un número aleatorio de una distribución uniforme: quiere que la probabilidad de obtener un número cualquiera sea la misma que la probabilidad de obtener cualquier otro número. La probabilidad de que aparezcan determinados números es uniforme, de ahí el nombre. A menudo también querrán que ese número aleatorio sea un número entero, lo que implica que quieren que proceda de una distribución uniforme discreta. Si quieren que ese número aleatorio sea cualquier decimal, implica que quieren que proceda de una distribución uniforme continua.

Un número muestreado a partir de una distribución uniforme continua que va de A a B puede tener cualquier valor entre esos dos extremos sin que ningún valor sea más o menos probable que cualquier otro valor posible.
- Si el intervalo es cerrado, el número muestreado puede tener cualquier valor entre A y B pero no A ni B exactamente. Esto se indica mediante corchetes: [A, B].
- Si el intervalo es abierto, el número muestreado puede tener cualquier valor entre A y B incluyendo A o B exactamente. Esto se indica mediante corchetes: (A,B).
- Intervalos semiabiertos - (A,B] o [A,B) son una mezcla: el número aleatorio podría ser igual al punto final del lado abierto, pero no al punto final del lado cerrado.

En Python, la función de SciPy toma muestras de un intervalo cerrado por defecto, mientras que la función uniform() de NumPy toma muestras de uno semiabierto - [A,B) por defecto. SciPy también nos da acceso a la función de densidad de probabilidad (PDF), la función de distribución acumulativa (CDF) y la función de punto porcentual (PPF), mientras que NumPy simplemente nos permite extraer números aleatorios.

### 4.8. Regresión <a class="anchor" id="t4_8"></a>

La regresión es un proceso estadístico para entender cómo una variable depende de otra variable. Por ejemplo, si se requiere entender cómo la edad de una persona afecta a su salario, se puede usar la regresión para encontrar una relación entre las dos variables. En términos simples, la regresión es una línea que se traza en un gráfico que muestra la relación entre dos variables. La línea se ajusta a los puntos de datos para mostrar la tendencia general entre las dos variables.

Tipos
- Modelo de regresión lineal simple: El análisis de regresión lineal simple es el más utilizado y el más sencillo de todos. Se trata de estudiar el efecto de una variable independiente sobre una única variable dependiente.
- Modelo de regresión lineal múltiple: El modelo cuenta con más de una variable independiente.
- Modelo de regresión no lineal: Hay ocasiones en las que la relación que puede darse entre variables independientes y la variable dependiente no tenga un desarrollo lineal, sino que tengan, por ejemplo, un crecimiento exponencial.


Los modelos de regresión incluyen los siguientes componentes:
- Los parámetros desconocidos, a menudo denotados como un escalar o vector β.
- Las variables independientes, que se observan en los datos y a menudo se denotan como un vector Xi ( donde i denota una fila de datos).
- La variable dependiente, que se observa en los datos y a menudo se denota mediante el escalar Yi
- Los términos de error, que no se observan directamente en los datos y a menudo se denotan utilizando el escalar ei

##### 4.8.1. Regresión lineal <a class="anchor" id="t4_8_1"></a>

La regresión lineal es un tipo de análisis predictivo que intenta predecir el valor de una variable dependiente con otra variable independiente. Estima los coeficientes de una ecuación lineal en la que intervienen una o varias variables independientes que predicen mejor la variable dependiente y ajusta una línea recta o superficie que reduce la variación entre los valores de salida previstos y los reales.


En el modelo de regresión lineal simple, dadas dos variables
- Y (dependiente, mayor coeficiente de variación)
- X (independiente, explicativa o predictora)
buscamos encontrar una función de X muy simple (lineal) que nos permita aproximar Y mediante: 

>Y´ = a + b X


Supuestos de la regresión lineal
Para que la regresión lineal tenga éxito, deben cumplirse cuatro supuestos. Entre ellos se incluyen:
- Este coeficiente sólo debe utilizarse para comparar variables cuantitativas y continuas.
- Debe existir una relación lineal entre las variables independiente y dependiente.
- Los residuos deben ser independientes, sin correlaciones entre ellos.
- Los residuos deben tener una varianza constante en cada nivel de x.
- Los residuos del modelo deben distribuirse normalmente.


 **¿Cómo identificar si dichas variables están relacionadas o no?**
 
 Para ellos podemos utilizar coeficientes, uno de los mas utilizados es el de correlación lineal de Pearson.

**Coeficiente de correlación lineal R**

El coeficiente de correlación, R, nos indica la fuerza y la dirección de la relación lineal entre x e y. Sin embargo, la fiabilidad del modelo lineal también depende del número de puntos de datos observados en la muestra. Tenemos que considerar conjuntamente el coeficiente de correlación r y el tamaño de la muestra n.


_Correlación_

La correlación trata de establecer la relación o dependencia que existe entre dos variables que intervienenen en una distribución.
- La correlación directa: Se da cuando al aumentar una de las variablesla la otra aumenta, por lo tanto, tendrá una pendiente positiva.
- La correlación inversa: Se da cuando al aumentar una de las variables la otra disminuye, tendrá una pendiente negativa.
- La correlación nula: Se da cuando no hay dependencia de ningún tipo entre las variables.


_Grado de correlación_

El grado de correlaciónindica la proximidad que hay entre los puntos de la nube de puntos,por lo que se asemejarán más o menos a una recta. Se pueden dar tres tipos:
- Correlación fuerte: La correlación será fuerte cuanta más cerca estén los puntos de la recta.
- Correlación débil: La correlación será fuerte cuanto más separados estén los puntos de la recta.
- Correlación nula: No se podrían asemejar los puntos a ninguna recta,no habría forma de discernir una recta entre los puntos.


Propiedades del coeficiente de correlación:
- Elcoeficienfe de correlación no varía al cambiar la escala de medición faly como pasaba conla covarianza.
- El signo del coeficiente de correlación es el mismo que el de la covarianza.
- El coeficiente de correlación lineal es un número real comprendido entre -1 y +1.
- Si el coeficiente de correlación lineal toma valores cercanos a -1 la correlación es fuerte e inversa. y será tanto más fuerte cuanto más se aproxime a r = -l.
- Si la coeficiente de correlación lineal toma valores cercanos a 1 la correlación es fuerte y directa. y será tanto más fuerte cuanto más se aproxime a r = l.
- Si el coeficiente de correlación lineal toma valores cercanos a 0<> la correlación es débil.


**Coeficiente de determinación R^2**

_Bondad de un ajuste_
- La dispersión del error residual será una fracción de la dispersión original de Y.
- Cuanto menor sea la dispersión del error residual mejor será el ajuste de regresión.

Propiedades del coeficiente de determinación R^2:
- R2 es una cantidad adimensional que sólo puede tomar valores en [0, 1]
- Cuando un ajuste es bueno, R2 será cercano a uno.
- Cuando un ajuste es malo R2 será cercano a cero.
- A R2 también se le denomina porcentaje de variabilidad explicado por el modelo de regresión.

### 4.9. Inferencia estadística <a class="anchor" id="t4_9"></a>

En muchas ocasiones, el problema no es determinar el valor de los parámetros de la población, sino verificar si un conjunto de afirmaciones sobre la población son ciertas o no, es decir, si deben aceptarse o no. El contraste de hipótesis trata de llegar a una
decisión correcta acerca de una hipótesis preestablecida.

#### 4.9.1. Introducción <a class="anchor" id="t4_9_1"></a>


Hipótesis estadística: Llamamos hipótesis estadística a una afirmación o conjetura con respecto a una o más poblaciones.

Aceptación y rechazo de una hipótesis: La aceptación de una hipótesis simplemente implica que los datos
no dan suficiente evidencia para rechazarla, mientras que el rechazo implica que la evidencia muestral la refuta (rechaza), o
bien que hay una pequeña probabilidad de obtener la información muestral observada cuando la hipótesis es verdadera.


_Hipótesis nula y alternativa_

- Llamamos hipótesis nula a cualquier hipótesis que deseemos contrastar y la notamos H0.
- El rechazo de H0 conduce a la aceptación de una hipótesis alternativa que denotaremos H1.
- El planteamiento de estas hipótesis ha de ser exhaustivo y excluyente. La hipótesis nula se enuncia de manera que incluya la igualdad.


_Contrastes unilaterales y bilaterales_

Según como enunciemos las hipótesis los contrastes pueden ser:
- Bilaterales si en H1 tomamos distinto frente una H0 de igualdad.
    - H0 : θ = θ0 
    - H1 : θ ≠ θ0
- Unilaterales si en H1 tomamos desigualdades estrictas frente a una H0 de desigualdades no estrictas.
    - H0 : θ ≥ θ0 ; H1 : θ < θ0 unilateral izquierdo
    - H0 : θ ≤ θ0 : H1 : θ > θ0 unilateral derecho

_Región crítica y nivel de significación_
- Región crítica
    - Valores ‘improbables’ si...
    - Es conocida antes de realizar el  experimento: resultados experimentales que refutarían H0
- Nivel de significación: α
    - Número pequeño: 1% , 5%
    - Fijado de antemano por el investigador
    - Es la probabilidad de rechazar H0 cuando es cierta

La posición de la región crítica depende de la hipótesis alternativa.

Estadístico de contraste: Llamamos estadístico de contraste a la fórmula muestral que utilizamos para tomar decisiones sobre la hipótesis nula. Si el valor del estadístico de contraste cae en la región crítica diremos “ el valor del estadístico es significativo ”

Valor p: Llamamos valor p, (p-valor o nivel crítico) al nivel de significación más bajo en el que el valor observado del estadístico de contraste es significativo. O sea: p = Probabilidad de obtener un estadístico al menos tan extremo como el observado bajo la hipótesis nula


**Procedimiento general**
- Describir las características de la variable en la población y de los datos muestrales.
- Formular las hipótesis estadísticas.
- Elegir y calcular el estadístico de contraste adecuado según las hipótesis y las características anteriores.
- Fijar el nivel de significación (α ). Para ese α, la zona de rechazo de H0 estará formada por todos los valores del estadístico cuya probabilidad de obtenerlos es muy pequeña bajo el supuesto de que H0 es cierta.


#### 4.9.2. Supuestos <a class="anchor" id="t4_9_2"></a>

Cuando se realiza una prueba de hipótesis de una media poblacional única μ utilizando una distribución t de Student (a menudo denominada prueba t), hay supuestos fundamentales que deben cumplirse para que la prueba funcione correctamente. 
- Sus datos deben ser una muestra aleatoria simple que proceda de una población que esté distribuida aproximadamente de forma normal. - Se utiliza la desviación típica de la muestra para aproximar la desviación típica de la población. 
- Tener en cuenta si el tamaño de la muestra es lo suficientemente grande, una prueba t funcionará incluso si la población no está distribuida de forma aproximadamente normal.

Cuando se realiza una prueba de hipótesis de una media poblacional única μ utilizando una distribución normal (a menudo denominada prueba z), se toma una muestra aleatoria simple de la población. 
- La población que está probando se distribuye normalmente o el tamaño de la muestra es suficientemente grande. 
- Se conoce el valor de la desviación típica de la población que, en realidad, rara vez se conoce.

Cuando se realiza una prueba de hipótesis de una única proporción poblacional p se toma una muestra aleatoria simple de la población. Debe cumplir las condiciones de una distribución binomial, que son las siguientes: 
- Hay un cierto número n de ensayos independientes, los resultados de cualquier ensayo son éxito o fracaso, y cada ensayo tiene la misma probabilidad de éxito p
- La forma de la distribución binomial debe ser similar a la de la distribución normal. Para ello, las cantidades np y nq deben ser mayores que cinco (np>5 y nq>5).

#### 4.9.3. Tipos de Errores <a class="anchor" id="t4_9_3"></a>

Cuando se realiza una prueba de hipótesis, hay cuatro resultados posibles que dependen de la veracidad o falsedad de la hipótesis nula H0 y de la decisión de rechazarla o no. Los resultados se resumen en la tabla siguiente:

<center><img src="../img/TiposDeErrores.jpeg"/></center>

Los cuatro resultados posibles de la tabla son los siguientes:
- La decisión es no rechazar H0 cuando H0 es verdadera (decisión correcta).
- La decisión es rechazar H0 cuando, de hecho, H0 es verdadera (decisión incorrecta conocida como error de tipo I).
- La decisión es no rechazar H0 cuando, de hecho, H0 es falsa (decisión incorrecta conocida como error de Tipo II).
- La decisión es rechazar H0 cuando H0 es falsa (decisión correcta cuya probabilidad se denomina Potencia de la Prueba).

Cada uno de los errores se produce con una probabilidad determinada. Las letras griegas α y β representan las probabilidades.
- α = probabilidad de un error de tipo I = P(Error de tipo I) = probabilidad de rechazar la hipótesis nula cuando la hipótesis nula es verdadera.
- β = probabilidad de un error de tipo II = P(error de tipo II) = probabilidad de no rechazar la hipótesis nula cuando la hipótesis nula es falsa.
- α y β deben ser lo más pequeñas posible porque son probabilidades de error. Rara vez son cero.

#### 4.9.4. Prueba T <a class="anchor" id="t4_9_4"></a>

> Se puede utilizar una prueba t cuando se han medido datos continuos (por ejemplo, la altura en cm o el peso en kg), en lugar de resultados binarios de aprobado/no aprobado o juicios cualitativos.

Para la prueba t de una muestra, los datos en cuestión serán una única muestra de mediciones y la pregunta estadística será, a grandes rasgos, si la media de esta muestra es la misma que la media de la población en su conjunto. En general, puede utilizarse si la población de datos de la que se extrae la muestra se distribuye normalmente, o si no se distribuye normalmente pero el tamaño de la muestra es grande.

**Generación de los datos**

```python
# Set a seed for the random number generator so we get the same random numbers each time
np.random.seed(20210725)

# Create fake data
mean = 100
standard_deviation = 5
sample_size = 20
x = np.random.normal(mean, standard_deviation, sample_size)

print([f'{x:.1f}' for x in sorted(x)])
```

**Estadística descriptiva**

```python
# Sample size
n = len(x)
# Mean
x_bar = np.mean(x)
# Sample standard deviation
s = np.std(x, ddof=1)
# Standard error of the mean
sem = s / np.sqrt(n)

print(f'Sample size = {n}, mean = {x_bar:.2f}, sample std deviation = {s:.2f}, std error of the mean = {sem:.2f}')
```

**Prueba t para una muestra**

El paquete Scipy tiene un subpaquete 'stats' que contiene la función ttest_1samp(). Esta función realiza pruebas t con una muestra. 

Tiene dos entradas obligatorias:
- La observación de la muestra (es decir, sus datos; los números que ha medido)
- La media de la población (es decir, el valor esperado que desea probar la media de la muestra en contra)

Esta función devuelve dos valores:
- El estadístico t, que puede considerarse el "resultado" de la prueba.
- El valor p, que puede considerarse como la probabilidad de que se haya producido ese resultado.

Podemos hacer una prueba unilateral en la que la hipótesis alternativa sea que la media poblacional real es mayor que la media poblacional dada utilizando el argumento de palabra clave alternative='greater':

```python
statistic, pvalue = stats.ttest_1samp(x, 98.5, alternative='greater')

print(f'p = {pvalue:5.3f}')
```

Estamos muy poco seguros (sólo un 2,4% de confianza) de que la hipótesis nula sea cierta y de que la media real de la población de la que se extrajo nuestra muestra sea inferior a 98,5 (lo cual es bueno, ¡la media real es 100!).

Por el contrario, podríamos haber hecho la prueba al revés, probando la hipótesis alternativa de que la media real es menor que la media de la población dada:

```python
statistic, pvalue = stats.ttest_1samp(x, 98.5, alternative='less')

print(f'p = {pvalue:5.3f}')
```

¡Ahora estamos muy seguros de que la hipótesis nula es correcta y de que la media poblacional verdadera es mayor que la media poblacional dada de 98,5!

**Prueba t de dos muestras**

Por defecto, la hipótesis nula utilizada por la función ttest_1samp() es que el valor esperado (media) de la muestra de observaciones independientes (los números de la primera entrada) es igual a la media de la población dada (la segunda entrada de la función). En otras palabras, es una prueba de dos muestras; la hipótesis nula es que la media de la población no es ni mayor ni menor que la media de la muestra.

Como estos datos son creado artificialmente, sabemos que el valor esperado de una muestra es 100 (porque hemos utilizado una distribución que tiene una media de 100), vamos a probar para ver si es igual a 98,5:

```python
# The population mean
popmean = 98.5
# Perform a two-sided one-sample t-test
statistic, pvalue = stats.ttest_1samp(x, popmean)

print(f'p-value = {pvalue:5.3f}')
```

Este valor p tan bajo nos indica que existe una pequeña probabilidad (menos del 5%) de que esta muestra se haya extraído aleatoriamente de una población con una media de 98,5. Si hubiéramos decidido utilizar un nivel de confianza del 95% como criterio de aceptación, habríamos rechazado la hipótesis nula tras ver este resultado.

**En resumen**

Prueba de dos muestras:
- Hipótesis nula: La media poblacional verdadera es igual a la media poblacional dada
- Hipótesis alternativa: La media poblacional real no es igual a la media poblacional dada
Prueba de una muestra, mayor que:
- Hipótesis nula: la media poblacional real no es mayor que la media poblacional dada
- Hipótesis alternativa: la media poblacional real es mayor que la media poblacional dada
Prueba de una muestra, menor que:
- Hipótesis nula: la media poblacional real no es menor que la media poblacional dada
- Hipótesis alternativa: la media poblacional real es menor que la media poblacional dada



Los resultados de las tres pruebas t que hemos realizado anteriormente están relacionados:
- El valor p de la prueba de dos caras era el doble que el de la prueba unilateral mayor que
- El valor p de la prueba unilateral mayor que más el valor p de la prueba unilateral menor que suman 1.
- El valor p de la prueba unilateral menor que sumado a la mitad del valor p de la prueba bilateral da un valor de 1.

La prueba de dos muestras pregunta si la media poblacional no es ni menor ni mayor que la media muestral, mientras que la prueba unilateral mayor que sólo pregunta si la media poblacional no es menor que la media muestral. Sólo hace la mitad de las preguntas, por lo que podemos estar doblemente seguros del resultado.

```python
# Two-sided test
statistic, pvalue_twosided = stats.ttest_1samp(x, 98.5)

# One-sided, greater than test
statistic, pvalue_greater = stats.ttest_1samp(x, 98.5, alternative='greater')

# One-sided, less than test
statistic, pvalue_less = stats.ttest_1samp(x, 98.5, alternative='less')

print(pvalue_twosided == pvalue_greater * 2)
print(pvalue_greater + pvalue_less == 1)
print(pvalue_less + pvalue_twosided / 2 == 1)
```

**Intervalos de confianza frente a la prueba t de una muestra**

Los intervalos de confianza del 95% para el valor de la media de la muestra pueden calcularse del siguiente modo:

```python
# Confidence level
alpha = 0.05  # 95% confidence
# Degrees of freedom
dof = n - 1
# Percent-point function (aka quantile function) of the t-distribution
t = stats.t.ppf(1 - (alpha / 2), dof)
# Margin of error
d = t * sem
# Intervals
upper_ci = x_bar + d
lower_ci = x_bar - d

print(f'Sample mean = {x_bar:.2f}, 95% CI [{lower_ci:.2f}, {upper_ci:.2f}]')
```

Esto implica que 98,5 está justo fuera del intervalo de números que esperamos que contenga la verdadera media poblacional y, por tanto, podemos afirmar, con un 95% de certeza, que la verdadera media poblacional y 98,5 no son iguales.

#### 4.9.5. Prueba Z <a class="anchor" id="t4_9_5"></a>

Se usa para comprobar si una muestra de datos, de los cuales una cierta proporción tiene alguna característica y quieres ver si esta proporción es significativamente diferente de la proporción que se espera.

Esta prueba, como todas las pruebas Z, consiste en calcular un estadístico Z. Suponiendo que los estadísticos Z se distribuyen normalmente, podemos calcular la improbabilidad de que la muestra haya producido el estadístico Z que ha producido. Más concretamente, el estadístico Z se compara con su distribución de referencia (la distribución normal estándar) para obtener un valor p.

**Carga de datos**

```python
import statsmodels.api as sm

# Load the data set as a dataframe-of-dataframes
data = sm.datasets.spector.load_pandas()
# Extract the complete dataset
df = data['data']
# Rename
df = df.rename(columns={
    'TUCE': 'test_score',
    'PSI': 'participated',
    'GRADE': 'grade_improved'
})

print(df.tail())
```

**Hipótesis**

Cuando se trabaja con datos categóricos hablamos de proporción, π en lugar de media, μ. Nos interesa saber si el 50% de los alumnos mejoraron o no su puntuación en el examen, es decir, si la verdadera proporción de alumnos que mejoraron es 0,5. Por tanto, nuestra hipótesis es:
- H0 : π=0.5
- H1 : π≠0.5

**Probar las hipótesis**
Si el tamaño de la muestra es grande (n>30) o se conoce la varianza de la población, podemos utilizar una prueba Z en lugar de una prueba T para la comprobación de la hipótesis. No conocemos la varianza de la población, pero el tamaño de nuestra muestra es superior a 30.

```python
from statsmodels.stats.proportion import proportions_ztest

# Number of successes
count = len(df[df['grade_improved'] == 1])
# Number of observations
nobs = len(df)
# Proportion under the null hypothesis
pi_0 = 0.5
# Perform a one-sample Z-test for a proportion
z_stat, p_value = proportions_ztest(count, nobs, value=pi_0, prop_var=pi_0)
# Proportion successful
pi = count / nobs

print(f'Proportion, π = {pi:.1%}; Z-statistic = {z_stat:.3f}; p = {p_value:.3f}')
```

**Comparación con la prueba de bondad de ajuste chi-cuadrado de Pearson**

Es importante señalar que este valor p es el mismo que el de la prueba de bondad de ajuste chi-cuadrado de Pearson:

```python
# Frequency of successful observations
f_obs = [count, nobs - count]
# Frequency of successful observations we expect under the null hypothesis
f_exp = [nobs * pi_0, nobs * pi_0]
# Perform a one-way chi-square test
chisq, p = stats.chisquare(f_obs, f_exp)
# Proportion of successful observations
pi = count / nobs

print(f'Proportion, π = {pi:.1%}; chi-squared, χ² = {chisq:.3f}; p = {p:.3f}')
```

**Intervalos de confianza**

```python
# Standard error of the proportion
se = np.sqrt((pi * (1 - pi)) / nobs)
# Significance level
alpha = 0.05
# Percent-point function (aka quantile function) of the normal distribution
z_critical = stats.norm.ppf(1 - (alpha / 2))
# Margin of error
d = z_critical * se
# Confidence interval
ci_lower = pi - d
ci_upper = pi + d

print(f'π = {pi:.1%} ± {d:.1%}')
```

**Interpretación**

- No rechazamos la hipótesis nula a favor de la alternativa, porque los datos observados no aportan pruebas en contra.
- Llegamos a la conclusión de que la proporción real de alumnos que mejoran su puntuación podría ser del 50%.
- Al generalizar a la población objetivo, tenemos un 95% de confianza en que el intervalo entre el 17,9% y el 50,8% contiene la verdadera proporción subyacente de alumnos que mejoraron su nota.
- Tenemos una confianza del 95% en que la proporción de nuestra muestra del 34,4% está dentro de los 16,5 pp (puntos porcentuales) de la verdadera proporción subyacente.

### 4.10. Análisis de varianza (ANOVA) <a class="anchor" id="t4_10"></a>

El análisis de varianza es un contraste para estudiar la igualdad de medias para más de dos muestras. La comparación entre las medias se efectúa en base al cálculo de la varianza entre las medias.

Podemos diferenciar dos tipos de análisis:
- Análisis univariante de la varianza, ANOVA (ANalisys Of Variance), en el que se estudian las diferencias entre grupos utilizando una única variable dependiente cuantitativa.
- Análisis multivariante de la varianza, MANOVA (Multivariante ANálisis Of VAriance) en el que se analizan las diferencias entre grupos a partir de múltiples variables dependientes cuantitativas.


El objetivo que pretende el ANOVA es el de determinar el efecto de una o más variables nominales u ordinales (factores) en una variable cuantitativa (Variable de respuesta).


Teóricamente se puede dividir la variabilidad del resultado de un experimento en dos partes:
- La originada por los factores que influyen directamente en el resultado, estudiados en sus distintos niveles o tratamientos.
- La producida por el resto de los factores cuya influencia en el resultado del experimento es desconocida o no controlable, que llamamos error experimental.

```python
# Load the dataset
wine = datasets.load_wine()

# Convert into a data frame

# Extract the data
data = pd.DataFrame(wine['data'], columns=wine['feature_names'])

# Extract the target
target = pd.DataFrame(wine['target'], columns=['cultivator'])

# Combine into one dataset
df = pd.concat([target, data], axis='columns')

# Trim the data
df = df[['cultivator', 'total_phenols']]

# Display the first 5 columns
print(df.head())
```

**Elegir una prueba estadística**

Al observar los datos trazados anteriormente surge una pregunta: ¿es diferente el contenido fenólico medio entre los tres cultivadores? Si lo es, implicaría que la calidad de los tres viticultores es diferente (suponiendo, por supuesto, que la calidad del vino sea objetiva).

**Paramétrica vs No paramétrica**

En primer lugar, debemos decidir si utilizamos estadísticas paramétricas o no paramétricas. Los estadísticos paramétricos se basan en suposiciones sobre la distribución de los datos (mientras que los no paramétricos no) y entre las razones para elegirlos se incluyen:
- Los datos se distribuyen normalmente.
- Los datos no se distribuyen normalmente, pero el tamaño de la muestra es grande:
    - n>20 si sólo tiene un grupo
    - n>15 en cada grupo si tiene de 2 a 9 grupos
    - n>20 en cada grupo si tiene de 10 a 12 grupos
- Las observaciones han de ser aleatorias e independientes (la dispersión de los datos en cada grupo es diferente)
- Los datos están sesgados (en otras palabras, la mediana es más representativa de la tendencia central que la media)
- Las varianzas de las subpoblaciones han de ser iguales.
- Es necesario aumentar la potencia estadística (las pruebas paramétricas suelen tener más potencia y, por tanto, más probabilidades de detectar una diferencia significativa cuando realmente existe).

Observando el gráfico, la dispersión en cada grupo es bastante consistente y nada está especialmente sesgado, aunque el tamaño de la muestra es grande y podría muy bien darse el caso de que los datos de cada grupo se aproximen a una distribución normal. En consecuencia, deberíamos optar por la estadística paramétrica y modificar nuestra pregunta de investigación para que se convierta en "¿es diferente el contenido fenólico medio?".

*Cuando se hace estadística paramétrica se debe utilizar la media. Si realiza estadísticas no paramétricas, utilice la mediana.

**Formular una hipótesis**

Para responder a la pregunta de investigación ("¿es diferente el contenido fenólico medio?") debemos formularla correctamente como una hipótesis nula con su correspondiente hipótesis alternativa:

- H0: Los verdaderos niveles medios de contenido fenólico son los mismos para los tres cultivadores.
- H1: Los verdaderos niveles medios de contenido fenólico no son los mismos para los tres cultivadores.

**Elija una prueba**

Ahora que tenemos hipótesis, podemos seguir un proceso de selección de pruebas estadísticas:
- Nuestros datos de contenido fenólico son continuos porque se componen de medidas que son números (no categorías discretas o rangos).
- Nos interesan las diferencias entre el contenido fenólico de los grupos de vino.
- En concreto, nos interesan las diferencias del contenido fenólico medio (no la varianza del contenido).
- Tenemos más de dos grupos porque hay tres cultivadores.

El siguiente diagrama de flujo nos indica que deberíamos utilizar ANOVA para comprobar nuestra hipótesis:

**ANOVA unidireccional con prueba F**

El modelo unifactorial se presenta cuando tenemos un solo factor, estudiado en sus distintos niveles, que influye sobre una variable respuesta que mide el resultado de un experimento y el resto de factores forman el error experimental.

Si los tamaños de todos los niveles son iguales decimos que estamos ante un modelo equilibrado y si no lo son, diremos que es un modelo no equilibrado.

Por tanto F compara la variabilidad debida a las diferencias entre los grupos o niveles con la variabilidad debida a las diferencias dentro de los niveles. Cuanto mayor sea el valor de F y menor su significación, más probabilidad de que existan diferencias significativas entre los grupos.

```python
# Samples
sample_0 = df[df['cultivator'] == 0]['total_phenols']
sample_1 = df[df['cultivator'] == 1]['total_phenols']
sample_2 = df[df['cultivator'] == 2]['total_phenols']
```

```python
# One-way ANOVA
statistic, pvalue = st.f_oneway(sample_0, sample_1, sample_2)

print(f'One-way ANOVA: s = {statistic:.2f}, p = {pvalue:.2e}')
```

**Interpretación de los resultados**

```python
def get_significance(p):
    "Devuelve la significación de un valor p como una cadena de estrellas."
    if p <= 0.001:
        return '***'
    elif p <= 0.01:
        return '**'
    elif p <= 0.05:
        return '*'
    elif p <= 0.1:
        return '.'
    else:
        return ''


def round_p_value(p):
    "Redondea un valor p pequeño para que sea legible."
    if p < 0.001:
        return '<0.001'
    else:
        return f'{p:5.3}'


p_rounded = round_p_value(pvalue)
significance = get_significance(pvalue)
print(f'P-value es {p_rounded} ({significance})')
```

### 5.1. Bases de datos relacionales <a class="anchor" id="t5_1"></a>

#### ¿Qué es un sistema de información?

Un SI es un conjunto de elementos relacionados entre sí de acuerdo a unas ciertas reglas, que aporta a la organización a la cual sirve y la que le marca las directrices de funcionamiento la información necesaria para el cumplimiento de sus fines. 

#### Ficheros vs Bases de datos

Si analizamos la situación (especialmente la situación pasada) de algunos sistemas de información, podremos encontrar sistemas en los que existe una proliferación de ficheros, específicos cada uno de ellos para una determinada aplicación. En estos ficheros, los datos se recogen varias veces y se encuentran repetidos en los distintos archivos. Esta redundancia, además de malgastar recursos, origina a menudo divergencias en los resultados. 

> “Una Base de Datos (BD) es una colección o depósito de datos integrados, almacenados en soporte secundario (no volátil) y con redundancia controlada”.


#### Tablas

El núcleo de una base de datos relacional son las tablas. Al igual que en la hoja de cálculo, una tabla es una colección de columnas y filas. La fila contiene los datos o la información con la que queremos trabajar, como el nombre de una ciudad o la cantidad de lluvia. Las columnas describen los datos que almacenan.


Elementos básicos
- Relación: Es la estructura básica del modelo relacional. Se representa mediante una tabla.
- Atributo: Representa las propiedades de la relación. Se representa mediante una columna.
- Dominio: Es el conjunto válido de valores que toma un atributo.
- Tupla: Es una ocurrencia de la relación. Se representa mediante una fila.


#### Las deficiencias de un enfoque de tabla única

Una de las inconvenientes son los datos duplicados, para resolver estos problemas tendríamos que modificar la estructura de nuestra tabla cada vez que hubiera un nuevo dato.

Por eso necesitamos múltiples tablas y relaciones. Al separar nuestros datos, podemos evitar la duplicación y tener más flexibilidad a la hora de trabajar con ellos.

Pero antes debemos averiguar cómo hacer referencia a cada dato. Necesitamos algún tipo de identificador, ID o (en términos técnicos de bases de datos) una clave primaria. Una clave primaria es un valor utilizado para identificar una fila específica de una tabla. Aunque podría basarse en un valor en sí (podríamos utilizar el nombre de la ciudad, por ejemplo), casi siempre debería ser un número u otro identificador. No queremos que el id cambie nunca, ya que rompería la relación. En la mayoría de los casos, la clave primaria o id será un número generado automáticamente.

#### Recuperación de datos

Si estamos utilizando una base de datos relacional como MySQL, SQL Server u Oracle, podemos utilizar un lenguaje llamado Lenguaje de Consulta Estructurado o SQL. SQL es un lenguaje estándar utilizado para recuperar y modificar datos en una base de datos relacional.

### 5.2. Bases de datos no relacionales <a class="anchor" id="t5_2"></a>

#### No-SQL

NoSQL es un término general que engloba las distintas formas de almacenar datos no relacionales y puede interpretarse como "no SQL", "no relacional" o "no sólo SQL".

> No-SQL: Los datos se almacenan de forma estructurada, pero no en filas y columnas, sino en documentos y éstos a su vez en colecciones de documentos.

Las bases de datos NoSQL se clasifican principalmente en cuatro tipos. Cada categoría tiene sus propios ventajas y limitaciones. Cada tipo se adapta mejor a un tipo de problema u otro por eso debemos seleccionar la mejor en cada caso.

Tipos de bases de datos NoSQL:
- Basadas en Pares Clave-Valor: Emparejan claves únicas, que son un identificador único asociado a un valor. Estos pares se almacenan utilizando una tabla hash con una función hash apropiada.
- Gráfico orientado a columnas: Organizan los datos en columnas y filas como una estructura de datos relacional, pero cada columna se divide en grupos denominados familia de columnas.
- Basada en gráfos: Describen las relaciones entre los datos y se representan como una colección de nodos y aristas. Las aristas representan la relación entre dos entidades Cada nodo y cada arista tienen propiedades que proporcionan información adicional sobre cada nodo y cada arista.
- Orientada a documentos: Almacena y recupera datos como un par clave-valor, pero la parte del valor se almacena como un documento. El documento se almacena en formatos JSON o XML. El valor es entendido por la BD y puede ser consultado.

#### La importancia de limpiar los datos

Dependiendo de su fuente, los datos brutos pueden contener algunas incoherencias que causarán problemas en el análisis y la modelización. En otras palabras, estos datos pueden clasificarse como "sucios" y necesitarán ser limpiados.

- Facilidad de uso y reutilización: Cuando los datos están correctamente organizados y normalizados es más fácil buscarlos, utilizarlos y compartirlos con otros.

Missing Data: Los datos que faltan pueden causar imprecisiones, así como resultados débiles o sesgados. A veces pueden resolverse "recargando" los datos, rellenando los valores que faltan con cálculos y código como Python, o simplemente eliminando el valor y los datos correspondientes.

### 6.2. Visualización por distribución (Frecuencia) <a class="anchor" id="t6_2"></a>

Una distribución de probabilidad es una descripción matemática de las posibles ocurrencias de un evento o variable aleatoria, junto con la probabilidad asociada a cada ocurrencia. En otras palabras, representa cómo se distribuyen las posibles valores de una variable aleatoria y la probabilidad de que ocurran esos valores.

Las distribuciones de probabilidad son fundamentales en estadísticas y teoría de probabilidades, y se utilizan para modelar y comprender el comportamiento de eventos aleatorios en diferentes contextos. Existen varios tipos de distribuciones de probabilidad:
- Distribución Uniforme: Cada valor tiene la misma probabilidad de ocurrir. Ejemplo: lanzar un dado justo.
- Distribución Binomial: Modela el número de éxitos en una serie de experimentos independientes, cada uno con dos resultados posibles. Ejemplo: éxito o fracaso en una serie de lanzamientos de moneda.
- Distribución Normal (Gaussiana): Tiene forma de campana y es ampliamente utilizada debido al teorema del límite central. Muchos fenómenos naturales se distribuyen de manera aproximadamente normal.
- Distribución Exponencial: Modela el tiempo entre eventos en un proceso de Poisson, donde los eventos ocurren a tasas constantes en el tiempo.
- Distribución Poisson: Modela la cantidad de eventos que ocurren en un intervalo de tiempo o espacio específico, cuando los eventos ocurren de manera rara pero aleatoria.
- Distribución Gamma: Generaliza la distribución exponencial y se utiliza en procesos donde el tiempo entre eventos no es constante.
- Distribución de Pareto: Modela fenómenos de cola larga, donde algunos eventos raros tienen un impacto significativo.

#### 7.1. Qué es la nube <a class="anchor" id="t7_1"></a>

La nube, o computación en nube, es la prestación de una amplia gama de servicios informáticos de pago por uso alojados en una infraestructura a través de Internet. Los servicios incluyen soluciones como almacenamiento, bases de datos, redes, software, análisis y servicios inteligentes.

Solemos diferenciar las nubes Pública, Privada e Híbrida de la siguiente manera:
- Nube pública: una nube pública es propiedad y está gestionada por un proveedor de servicios en la nube externo que ofrece sus recursos informáticos a través de Internet al público.
- Nube privada: se refiere a los recursos de computación en nube utilizados exclusivamente por una sola empresa u organización, con servicios e infraestructura mantenidos en una red privada.
- Nube híbrida: la nube híbrida es un sistema que combina nubes públicas y privadas. Los usuarios optan por un centro de datos local, al tiempo que permiten que los datos y las aplicaciones se ejecuten en una o varias nubes públicas.

La mayoría de los servicios de computación en nube se dividen en tres categorías: Infraestructura como servicio (IaaS), Plataforma como servicio (PaaS) y Software como servicio (SaaS).
- Infraestructura como servicio (IaaS): los usuarios alquilan una infraestructura informática como servidores y máquinas virtuales (VM), almacenamiento, redes y sistemas operativos.
- Plataforma como servicio (PaaS): los usuarios alquilan un entorno para desarrollar, probar, entregar y gestionar aplicaciones de software. Los usuarios no tienen que preocuparse de configurar o gestionar la infraestructura subyacente de servidores, almacenamiento, red y bases de datos necesaria para el desarrollo.
- Software como servicio (SaaS): los usuarios obtienen acceso a aplicaciones de software a través de Internet, bajo demanda y normalmente por suscripción. Los usuarios no tienen que preocuparse de alojar y gestionar la aplicación de software, la infraestructura subyacente o el mantenimiento, como las actualizaciones de software y los parches de seguridad.


#### 7.2. Por qué elegir la nube <a class="anchor" id="t7_2"></a>

Algunas razones incluyen las siguientes:
- Innovación: Puedes potenciar tus aplicaciones integrando servicios innovadores creados por proveedores de Cloud directamente en tus apps.
- Flexibilidad: Sólo paga por los servicios que necesita y puede elegir entre una amplia gama de servicios. Suele pagar sobre la marcha y adaptar sus servicios en función de la evolución de sus necesidades.
- Presupuesto: No tiene que hacer inversiones iniciales para adquirir hardware y software, instalar y gestionar centros de datos in situ, y sólo tiene que pagar por lo que utiliza.
- Escalabilidad: Sus recursos pueden escalarse en función de las necesidades de su proyecto, lo que significa que sus aplicaciones pueden utilizar más o menos potencia informática, almacenamiento y ancho de banda, adaptándose a los factores externos en cada momento.
- Productividad: Puedes centrarte en tu negocio en lugar de dedicar tiempo a tareas que puede gestionar otra persona, como la gestión de centros de datos.
- Fiabilidad: El Cloud Computing ofrece varias formas de realizar copias de seguridad continuas de tus datos y puedes establecer planes de recuperación ante desastres para mantener tu negocio y tus servicios en marcha, incluso en tiempos de crisis.
- Seguridad: Puede beneficiarse de políticas, tecnologías y controles que refuerzan la seguridad de su proyecto.
